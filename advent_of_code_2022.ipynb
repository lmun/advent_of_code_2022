```python
file_name = 'day7.txt'
with open(file_name, 'r') as f:
    t = f.read()

lines = t.split('\n')

todos = []
class Aux:
    def __init__(self, parent, dirname):
        self.files = set()
        self.dirs = {}
        self.parent = parent
        self.name = dirname
    def get_sizes(self):
        res = 0
        for dir in self.dirs.values():
            for f, n in dir.get_sizes():
                # yield f, n
                res += f
        res += sum(a[0] for a in self.files)
        # print(self.name, res)
        todos.append((self.name, res))
        yield res, self.name

root = Aux(None, '/')
ite = root
i = 0
while i < len(lines):
    line = lines[i]
    if line[0] == '$':
        if line.startswith('$ cd '):
            if line[5:] == '/':
                ite = root
            elif line[5:] == '..':
                ite = ite.parent
            else:
                ite = ite.dirs[line[5:]]
        if line == '$ ls':
            while i+1 < len(lines) and not lines[i+1].startswith('$'):
                nline = lines[i+1]
                if nline.startswith('dir'):
                    ndir = nline[4:]
                    if ndir not in ite.dirs:
                        ite.dirs[ndir] = Aux(ite, ite.name + ndir+'/')
                else:
                    s, n = int(nline[:nline.index(' ')]), nline[nline.index(' '):]
                    # print(s,n)
                    ite.files.add((s,n))
                i += 1
    i += 1
falta = 0
for x in root.get_sizes():
    print(x)
    falta = x[0]-40000000
    # pass
print(falta)
res = 0
for n,s in todos:
    if s<=100000:
        res+=s
for n,s in sorted(todos, key=lambda x: x[1]):
    if s >= falta:
        print(n,s)
        break
print(res)
```

```python
from ast import literal_eval
from heapq import heappush, heappop
from itertools import zip_longest
file_name = 'day13.txt'
# file_name = 'day13_test.txt'
with open(file_name, 'r') as f:
    t = f.read()
pairs = t.split('\n\n')

def get_order(l,r):
    if l is None:
        if r is None:
            return 0
        return 1
    if r is None:
        return -1
    if type(l) is int and  type(r) is int:
        return r - l
    if type(l) is list and type(r) is list:
        for a,b in zip_longest(l,r):
            dab = get_order(a,b)
            if dab != 0:
                return dab
    if type(l) is int:
        return get_order([l], r)
    if type(r) is int:
        return get_order(l, [r])
    return 0
res = 0

do,se = [[2]], [[6]]
ma,me = 1, 2
for i, pair in enumerate(pairs, 1):
    left,right = map(literal_eval, pair.split('\n'))
    if get_order(left, do) > 0:
        # print(left)
        ma += 1
    if get_order(right, do) > 0:
        # print(right)
        ma += 1
    if get_order(left, se) > 0:
        me +=1
    if get_order(right, se) > 0:
        me += 1
    if get_order(left, right) > 0:
        res += i
        pass
print(res)
print(ma,me)
print(ma*me)
```
